```python
movie_ids = []
movie_titles = []
counter = 0



for movie in titles:
    try:
        if counter <= 40:
            URL = 'https://api.themoviedb.org/3/search/movie?api_key={}&query={}'.format(api_key, movie)
            response = requests.get(URL)
            movie_ids.append(response.json()['results'][0]['id'])
            counter+=1
            movie_titles.append(movie)
        else: 
            print('Rate Limit Reached')
            sleep(10)
            counter = 0
    except:
        pass
```
